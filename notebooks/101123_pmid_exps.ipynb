```python
from Bio import Entrez

# Always provide your email address
Entrez.email = "julie.r.vaughn@gmail.com"

def search(query):
    handle = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax='2',
                            retmode='xml', 
                            term=query)
    results = Entrez.read(handle)
    return results

def fetch_details(id_list):
    ids = ','.join(id_list)
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results
```

```python

# Example usage
results = search('epigenetic clocks and female fertility')
id_list = results['IdList']
papers = fetch_details(id_list)
for i, paper in enumerate(papers['PubmedArticle']):
    print(f"{i+1}) {paper['MedlineCitation']['Article']['ArticleTitle']}")
    # Check for full text links
    if 'OtherAbstract' in paper['PubmedData']:
        for other_abstract in paper['PubmedData']['OtherAbstract']:
            if other_abstract['@Type'] == 'Publisher':
                print(" Full text link: ", other_abstract['AbstractText'])
```

```python
from Bio import Entrez

Entrez.email = "your_email@example.com"  # Always provide your email

def fetch_details(pmid):
    handle = Entrez.efetch(db='pubmed', retmode='xml', id=pmid)
    results = Entrez.read(handle)
    return results

def get_full_text_link(pmid):
    paper = fetch_details(pmid)
    article = paper['PubmedArticle'][0]
    article_data = article['PubmedData']
    full_text_link = None

    # Check for links to full text
    if 'ArticleIdList' in article_data:
        article_ids = article_data['ArticleIdList']
        for article_id in article_ids:
            if article_id.attributes['IdType'] == 'doi':
                doi = article_id
                full_text_link = f"https://doi.org/{doi}"
            elif article_id.attributes['IdType'] == 'pmc':
                pmc_id = article_id
                full_text_link = f"https://www.ncbi.nlm.nih.gov/pmc/articles/{pmc_id}/"

    return full_text_link

# Example usage
pmid = "37025178"  # Replace with a valid PMID
full_text_url = get_full_text_link(pmid)
if full_text_url:
    print(f"Full text available at: {full_text_url}")
else:
    print("Full text link not found.")
```

```python
import requests
response = requests.get(full_text_url)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
text = soup.get_text()
print(text)
```
