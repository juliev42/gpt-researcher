```python
from tavily import TavilyClient

from dotenv import load_dotenv
import os
load_dotenv()
```

```python
tavily = TavilyClient(api_key=os.environ['TAVILY_API_KEY'])
```

```python
response = tavily.search(query="What are the risk factors for osteoporosis in women?", max_results = 2, include_domains = ["https://www.ncbi.nlm.nih.gov/"], search_depth="advanced", include_raw_content=True)
```
